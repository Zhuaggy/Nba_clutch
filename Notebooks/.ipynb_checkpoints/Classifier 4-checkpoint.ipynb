```python
total_df = final_df[['PERSON_ID','SEASON_ID','FG_PCT', 'FG3_PCT', 'FT_PCT', 'OREB', 'DREB', 'REB', 'AST', 'TOV',
       'STL', 'BLK', '4Q_FG_PCT', '4Q_FG3_PCT', '4Q_FT_PCT', '4Q_OREB',
       '4Q_DREB', '4Q_REB', '4Q_AST', '4Q_TOV', '4Q_STL', '4Q_BLK',
        'GP', 'GS', 'AVG_MIN',
       'AGE', 'SEASON', 'RA_FGA', 'RA_FG_PCT', 'P_FGA', 'P_FG_PCT', 'MR_FGA',
       'MR_FG_PCT', 'LC_FGA', 'LC_FG_PCT', 'RC_FGA', 'RC_FG_PCT', 'ATB_FGA',
       'ATB_FG_PCT', 'BC_FGA', 'BC_FG_PCT', 'total_shot_attempts', 'RA_FQ',
       'P_FQ', 'MR_FQ', 'LC_FQ', 'RC_FQ', 'ATB_FQ', 'BC_FQ', '0D_FQ',
       '0D_FG_PCT', '1D_FQ', '1D_FG_PCT', '2D_FQ', '2D_FG_PCT', '3D_FQ',
       '3D_FG_PCT', 'VT_FQ', 'VT_FG_PCT', 'T_FQ', 'T_FG_PCT', 'O_FQ',
       'O_FG_PCT', 'WO_FQ', 'WO_FG_PCT', 'CT_FG_PCT', 'CT_FGA']]
total_df = total_df.dropna()
total_df = total_df[total_df['CT_FGA'] > 0]
total_df['CT_FG_PCT_DIFF'] = total_df['CT_FG_PCT'] - total_df['FG_PCT']
total_df['Choker'] = total_df['CT_FG_PCT_DIFF'].apply(to_category_1)
```

```python
len(total_df)
```

```python
target = total_df['Choker']

features = total_df[['PERSON_ID','FG_PCT', 'FG3_PCT', 'FT_PCT', 'OREB', 'DREB', 'REB', 'AST', 'TOV',
       'STL', 'BLK', '4Q_FG_PCT', '4Q_FG3_PCT', '4Q_FT_PCT', '4Q_OREB',
       '4Q_DREB', '4Q_REB', '4Q_AST', '4Q_TOV', '4Q_STL', '4Q_BLK',
        'GP', 'GS', 'AVG_MIN',
       'AGE', 'RA_FGA', 'RA_FG_PCT', 'P_FGA', 'P_FG_PCT', 'MR_FGA',
       'MR_FG_PCT', 'LC_FGA', 'LC_FG_PCT', 'RC_FGA', 'RC_FG_PCT', 'ATB_FGA',
       'ATB_FG_PCT', 'BC_FGA', 'BC_FG_PCT', 'total_shot_attempts', 'RA_FQ',
       'P_FQ', 'MR_FQ', 'LC_FQ', 'RC_FQ', 'ATB_FQ', 'BC_FQ', '0D_FQ',
       '0D_FG_PCT', '1D_FQ', '1D_FG_PCT', '2D_FQ', '2D_FG_PCT', '3D_FQ',
       '3D_FG_PCT', 'VT_FQ', 'VT_FG_PCT', 'T_FQ', 'T_FG_PCT', 'O_FQ',
       'O_FG_PCT', 'WO_FQ', 'WO_FG_PCT','CT_FGA']]
X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.33, random_state=0)
```

```python
estimator = RandomForestClassifier(random_state=0, n_estimators=600)
estimator.fit(X_train, y_train)
```

```python
imp_df = pd.DataFrame()
imp_df['importance'] = estimator.feature_importances_
imp_df['feature'] = features.columns
```

```python
imp_df.sort_values(by = 'importance', ascending = False)
```

```python
df_results = pd.DataFrame()
df_results['prediction'] = estimator.predict(X_test)
df_results['actual'] = y_test.values
df_results['abs_error'] = abs(df_results['actual'] - df_results['prediction'])
df_results['pct_error'] = df_results['abs_error']/df_results['actual']
```

```python
def to_category_1(x):
    if x <= 0:
        return (1)
    else:
        return (0)

def to_category_2(x):
    if x <= 0.5:
        return (1)
    else:
        return (0)
```

```python
df_results['abs_error'] = abs(df_results['actual'] - df_results['prediction'])
df_results['pct_error'] = df_results['abs_error']/df_results['actual']
print('abs cat error: ' + str(df_results['abs_error'].mean()))
print('number of actual clutch correct: ' + str(1 - df_results[df_results['actual'] == 1]['abs_error'].mean()))
print('number of chokers correct: ' + str(1 - df_results[df_results['actual'] == 0]['abs_error'].mean()))
print('number of actual clutch: ' + str(df_results['actual'].sum()))
print('number of predicted clutch: ' + str(df_results['prediction'].sum()))
```
