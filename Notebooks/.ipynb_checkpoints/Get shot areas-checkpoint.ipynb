```python
%matplotlib inline
import nba_py as nba
from nba_py import player
import requests
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from tqdm import tqdm
import math
```

```python
comb_df = pd.read_csv('july18_data.csv')
```

```python
#1998-99 -> 2016-17
sa_years = ['1998-99','1999-00','2000-01','2001-02','2002-03','2003-04','2004-05','2005-06','2006-07',
           '2007-08','2008-09','2009-10','2010-11','2011-12','2012-13','2013-14','2014-15','2015-16','2016-17']
temp_list = [['PERSON_ID','SEASON','RA_FGA','RA_FG_PCT','P_FGA','P_FG_PCT','MR_FGA','MR_FG_PCT','LC_FGA','LC_FG_PCT','RC_FGA','RC_FG_PCT',
            'ATB_FGA','ATB_FG_PCT','BC_FGA','BC_FG_PCT']]
for year in tqdm(sa_years):
    cur_df = comb_df[comb_df['GROUP_VALUE'] == year]
    for pid in tqdm(cur_df['PERSON_ID'].unique()):
        pss = player.PlayerShootingSplits(pid, season=year).shot_areas()
        cur_list = []
        if pss.empty == False and len(pss) == 7:
            cur_list.append(pid)
            cur_list.append(year)
            for i in range(7):
                cur_list.append(pss.iloc[i][3])
                cur_list.append(pss.iloc[i][4])
            temp_list.append(cur_list)
```
