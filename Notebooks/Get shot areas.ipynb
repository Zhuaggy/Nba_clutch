```python
%matplotlib inline
import nba_py as nba
from nba_py import player
import requests
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from tqdm import tqdm
import math
```

```python
comb_df = pd.read_csv('july18_data.csv')
```

```python
#1998-99 -> 2016-17
sa_years = ['2013-14','2014-15','2015-16','2016-17']
temp_list = [['PERSON_ID','SEASON','0D_FQ','0D_FG_PCT','1D_FQ','1D_FG_PCT','2D_FQ','2D_FG_PCT','3D_FQ','3D_FG_PCT']]
for year in sa_years:
    cur_df = comb_df[comb_df['GROUP_VALUE'] == year]
    for pid in tqdm(cur_df['PERSON_ID'].unique()):
        pss = player.PlayerShotTracking(pid, season=year).dribble_shooting()
        cur_list = []
        if pss.empty == False and len(pss) >= 4:
            cur_list.append(pid)
            cur_list.append(year)
            for i in range(4):
                cur_list.append(pss.iloc[i][6])
                cur_list.append(pss.iloc[i][9])
            temp_list.append(cur_list)
```

```python
pd.DataFrame(temp_list).to_csv('dribble_july18.csv')
```

```python
drib_df = pd.read_csv('dribble_july18.csv')
def_df = pd.read_csv('defender_july18.csv')
sa_df = pd.read_csv('shot_areas.csv')
del sa_df['Unnamed: 0']
del drib_df['Unnamed: 0']
del def_df['Unnamed: 0']
```

```python
drib_df.columns = drib_df.iloc[0]
drib_df = drib_df.reindex(drib_df.index.drop(0))
def_df.columns = def_df.iloc[0]
def_df = def_df.reindex(def_df.index.drop(0))
sa_df.columns = sa_df.iloc[0]
sa_df = sa_df.reindex(sa_df.index.drop(0))
```

```python
sa_df.head()
```

```python
sa_df['total_shot_attempts'] = sa_df['RA_FGA'] + sa_df['P_FGA'] + sa_df['MR_FGA'] + sa_df['LC_FGA'] + sa_df['RC_FGA'] + sa_df['ATB_FGA'] + sa_df['BC_FGA']
```

```python
sa_df['RA_FGA'] = sa_df['RA_FGA'].astype(float)
sa_df['P_FGA'] = sa_df['P_FGA'].astype(float)
sa_df['MR_FGA'] = sa_df['MR_FGA'].astype(float)
sa_df['LC_FGA'] = sa_df['LC_FGA'].astype(float)
sa_df['RC_FGA'] = sa_df['RC_FGA'].astype(float)
sa_df['ATB_FGA'] = sa_df['ATB_FGA'].astype(float)
sa_df['BC_FGA'] = sa_df['BC_FGA'].astype(float)


sa_df['total_shot_attempts'] = sa_df['RA_FGA'] + sa_df['P_FGA'] + sa_df['MR_FGA'] + sa_df['LC_FGA'] + sa_df['RC_FGA'] + sa_df['ATB_FGA'] + sa_df['BC_FGA']


sa_df['RA_FQ'] = sa_df['RA_FGA']/sa_df['total_shot_attempts']
sa_df['P_FQ'] = sa_df['P_FGA']/sa_df['total_shot_attempts']
sa_df['MR_FQ'] = sa_df['MR_FGA']/sa_df['total_shot_attempts']
sa_df['LC_FQ'] = sa_df['LC_FGA']/sa_df['total_shot_attempts']
sa_df['RC_FQ'] = sa_df['RC_FGA']/sa_df['total_shot_attempts']
sa_df['ATB_FQ'] = sa_df['ATB_FGA']/sa_df['total_shot_attempts']
sa_df['BC_FQ'] = sa_df['BC_FGA']/sa_df['total_shot_attempts']
```
